```python
import string
letters = [i for i in string.ascii_uppercase]

blocks = []
with open('day5.txt') as f:
    f = [i.strip() for i in f]
    for idx,i in enumerate(f):
#         print(i)
        blocks.append(i)
        if idx == 7:
            break
blocks_rev = blocks[::-1]

letter_positions = {}
column = 1
for idx,i in enumerate(blocks_rev[0]):
    if i in letters:
        letter_positions[idx] = column
        column+=1
        
board_dict = {}
for idx,i in enumerate(blocks_rev):
    if idx == 0:
        for j in letter_positions:
            if i[j] in letters:
                board_dict[letter_positions[j]] = list(i[j])
    else:
        for j in letter_positions:
            try:
                if i[j] in letters:
                    board_dict[letter_positions[j]].append(i[j])
            except:
                _

def move(num,fr,to):
    x = board_dict[fr][-num:]
    y = x[::-1]
#     print(x)
    del(board_dict[fr][-num:])
    for i in y:
        board_dict[to].append(i)
#     print(x)
#     print(board_dict)
    return x
commands = []
import re
with open('day5.txt') as f:
    f = [i.strip() for i in f]
    for idx,i in enumerate(f[10:]):
#         print(i)
        commands.append(i)
for i in commands:
    x = re.findall('\s[0-9]+',i)
    move(int(x[0]),int(x[1]),int(x[2]))

result = []
for i in board_dict:
    result.append(board_dict[i][-1])
print(*result,sep="")
```
